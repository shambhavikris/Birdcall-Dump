```
#this notebook is for creating TF records, 2 for each class, from .npz files which are already separated into train and test
```

```
import tensorflow as tf
import numpy as np
import pandas as pd
from numpy import asarray
from numpy import savez_compressed
import os
from numpy import load
```

```
#first lets get mapping done so that labels can be assigned
```

```
df = pd.read_csv('/content/drive/My Drive/train.csv')
ser = df['ebird_code']
lis = ser.tolist()
classes = np.unique(lis)
print(len(classes)) #264
dict_class = {}
dict_class = { classes[i] : i for i in range(0, len(classes))}
print(dict_class['amebit']) #2
print(dict_class['yerwar']) #262
```

```
# next, the functions for tf record creation
```

```
def _bytes_feature(value):
    """Returns a bytes_list from a string / byte."""
    # If the value is an eager tensor BytesList won't unpack a string from an EagerTensor.
    if isinstance(value, type(tf.constant(0))):
        value = value.numpy() 
    return tf.train.Feature(bytes_list=tf.train.BytesList(value=[value]))

def _float_feature(value):
    """Returns a float_list from a float / double."""
    return tf.train.Feature(float_list=tf.train.FloatList(value=[value]))

def _int64_feature(value):
    """Returns an int64_list from a bool / enum / int / uint."""
    return tf.train.Feature(int64_list=tf.train.Int64List(value=[value]))
```

```
def serialize_example(feature0, feature1):
  feature = {
      'feature0': _bytes_feature(feature0),
      'feature1': _int64_feature(feature1),
  }
# # Create a Features message using tf.train.Example.
  example_proto = tf.train.Example(features=tf.train.Features(feature=feature))
  return example_proto.SerializeToString()
```

```
train_dir = '/content/drive/My Drive/lala1/output-npz/train-npz'
test_dir = '/content/drive/My Drive/lala1/output-npz/test-npz'
```

```
train_output = '/content/drive/My Drive/lala1/train-tfr'
test_output = '/content/drive/My Drive/lala1/test-tfr'
```

```
npz_train = os.listdir(train_dir)
npz_test = os.listdir(test_dir)
```

```
#loop through all classes for train
num = 0

for bird in npz_train:
    num = num+1
    label = dict_class[bird.split('_')[0]]

    bird_path = os.path.join(train_dir, bird)
    dict_data = load(bird_path)
    data = dict_data['arr_0']
    data = np.transpose(data)
    print(data.shape)

    output_file = train_output + '/' + bird.split('_')[0] + '_train.tfrecord'
    print('class tf record for ', bird.split('_')[0], ' train is:', output_file)

    options = tf.io.TFRecordOptions(compression_type="GZIP")

    #loop through the array in the .npz file and add one by one to a tf record
    with tf.io.TFRecordWriter(output_file, options=options) as writer:

        for i in data:
            i = tf.convert_to_tensor(i)
            i = tf.expand_dims(i, -1)
            i.set_shape([900000,1])
            serialized_example = serialize_example(tf.io.serialize_tensor(i), label)
            writer.write(serialized_example)
```

```
print('done with train')
print(num) # 96
print('restart session to clear RAM before test loop is run')
```

```
#loop through all classes for test
num = 0 

for bird in npz_test:
    num = num + 1
    label = dict_class[bird.split('_')[0]]

    bird_path = os.path.join(test_dir, bird)
    dict_data = load(bird_path)
    data = dict_data['arr_0']
    data = np.transpose(data)
    print(data.shape)

    output_file = test_output + '/' + bird.split('_')[0] + '_test.tfrecord'
    print('class tf record for ', bird.split('_')[0], ' test is:', output_file)

    options = tf.io.TFRecordOptions(compression_type="GZIP")

    #loop through the array in the .npz file and add one by one to a tf record
    with tf.io.TFRecordWriter(output_file, options=options) as writer:

        for i in data:
            i = tf.convert_to_tensor(i)
            i = tf.expand_dims(i, -1)
            i.set_shape([900000,1])
            serialized_example = serialize_example(tf.io.serialize_tensor(i), label)
            writer.write(serialized_example)
```
