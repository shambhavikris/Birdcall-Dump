```
import pandas as pd
import numpy as np
from numpy import asarray
from numpy import savez_compressed
import os
import tensorflow as tf
import pydub
```

```
dir = '/content/drive/My Drive/lala1/train_audio/dataupload residual'
output_dir = '/content/drive/My Drive/lala1/output-npz'
ms_of_data = 30000*30

for d in os.listdir(dir):
  folder = os.path.join(dir, d)
  #print('folder', folder)
  #num1 = 1

  for subset in os.listdir(folder): #test or train
      
      path_tt = os.path.join(folder, subset)
      flag = 0 #create new array

      for f in os.listdir(path_tt):
        # each .wav file in either test or train
        fn = os.path.join(path_tt, f)
        print('fn', fn)
        if not fn.endswith('wav'):
          print('skip, not wav')
          continue

        x = pydub.AudioSegment.from_file(file=fn)
        # os.remove(fn)
        x = x.set_frame_rate(30000)
        x = x.set_sample_width(1)
        x = np.array(x.get_array_of_samples(), dtype=np.float32)
        x = np.trim_zeros(x)
        num = len(x)//ms_of_data
        if num==0:
          x = np.pad(x, (0, ms_of_data - len(x)))
        x = x[0:(len(x)//ms_of_data)*ms_of_data]
        # print(str(num1), 'windows', str(len(x)//ms_of_data))
        x = np.array(np.split(x, len(x)//ms_of_data))
        print('shape of x', x.shape)
        for i in x: ##for each window, appended to total
          i = np.expand_dims(i, axis=-1)
          print('shape of i', i.shape)
  
          if flag == 0:
            flag = 1 #first done
            subset_total = i
          else:
            subset_total = np.concatenate((subset_total, i), axis=1)

      #done with test or train subset of a class, save the total array
      filename = output_dir + '/' + d + '_' + subset + '.npz'
      print('saving from ', d, 'class to', filename)
      savez_compressed(filename, subset_total)
      print('shape of this ^ is ', subset_total.shape)
      
```

```
print('done data residual')
```
