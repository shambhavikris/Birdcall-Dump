```
# this notebook is for separating the test and train .npz files from output-npz to 2 folders, train-split and test-split
```

```
import os
import shutil
```

```
dir = '/content/drive/My Drive/lala1/output-npz'
```

```
output_train = '/content/drive/My Drive/lala1/train_audio/train-npz'
output_test = '/content/drive/My Drive/lala1/train_audio/test-npz'
```

```
t1 = 0
t2 = 0
for f in os.listdir(dir):
    src = os.path.join(dir, f)
    if f.find('train') == -1:
        #it is a test file
        shutil.move(src, output_test)
        print(f, ' moved to ', output_test)
        t1 = t1+1
    else:
        shutil.move(src, output_train)
        print(f, ' moved to ', output_train)
        t2 = t2+1
```

```
print('moved')
print(t1)
print(t2)
```
